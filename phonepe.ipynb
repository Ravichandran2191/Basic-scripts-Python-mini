```python
import pandas as pd
import json
import os
import csv
```

```python
import csv
```

```python
path='/content/pulse/data/aggregated/transaction/country/india/state/'
Agg_State_list=os.listdir(path)
Agg_State_list
```

```python
clm={'state':[],'year':[],'quarter':[],'transaction_type':[],'transaction_count':[],'transaction_amount':[]}
for i in Agg_State_list:
  p_i=path+i+'/'
  Agg_yr=os.listdir(p_i)
  for j in Agg_yr:
    p_j=p_i+j+'/'
    Agg_yr_list=os.listdir(p_j)
    for k in Agg_yr_list:
        p_k=p_j+k
        data=open(p_k,'r')
        d=json.load(data)
        for z in d['data']['transactionData']:
          Name=z['name']
          count=z['paymentInstruments'][0]['count']
          amount=z['paymentInstruments'][0]['amount']
          clm['transaction_type'].append(Name)
          clm['transaction_count'].append(count)
          clm['transaction_amount'].append(amount)
          clm['state'].append(i)
          clm['year'].append(j)
          clm['quarter'].append(int(k.strip('.json')))
df1=pd.DataFrame(clm)
df1
```

```python
df1.to_csv('z.csv')
```

```python
path2='/content/pulse/data/aggregated/user/country/india/state/'
Agg_state2=os.listdir(path2)
Agg_state2
```

```python
clm2={'state':[],'year':[],'quater':[],'brand':[],'count':[],'percentage':[]}
for x in Agg_state2:
  x_i=path2+x+'/'
  Agg_yr2=os.listdir(x_i)
  for l in Agg_yr2:
    l_i=x_i+l+'/'
    Agg_js=os.listdir(l_i)
    for js in Agg_js:
      j_js=l_i+js
      data2=open(j_js,'r')
      d2=json.load(data2)
      if d2['data']['usersByDevice'] is not None:
        for a in d2['data']['usersByDevice']:
          brand=a['brand']
          count=a['count']
          percentage=a['percentage']
          clm2['brand'].append(brand)
          clm2['count'].append(count)
          clm2['percentage'].append(percentage)
          clm2['state'].append(x)
          clm2['year'].append(l)
          clm2['quater'].append(int(js.strip('.json')))
df2=pd.DataFrame(clm2)
df2
```

```python
df2.to_csv('a.csv')
```

```python
path3='/content/pulse/data/map/transaction/hover/country/india/state/'
Agg_map_state=os.listdir(path3)
Agg_map_state
```

```python
clm3={'stste':[],'year':[],'quater':[],'metric_count':[],'metric_amount':[]}
for m in Agg_map_state:
  m_s=path3+m+'/'
  Agg_map_yr=os.listdir(m_s)
  Agg_map_yr
  for my in Agg_map_yr:
    my_y=m_s+my+'/'
    # print(my_y)
    Agg_map_js=os.listdir(my_y)
    Agg_map_js
    for j in Agg_map_js:
      j_s=my_y+j
      # print(j_s)
      data3=open(j_s,'r')
      d3=json.load(data3)
      for s in d3['data']['hoverDataList']:
        count=s['metric'][0]['count']
        amount=s['metric'][0]['amount']
        clm3['metric_count'].append(count)
        clm3['metric_amount'].append(amount)
        clm3['stste'].append(m)
        clm3['year'].append(my)
        clm3['quater'].append(int(j.strip('.json')))
df3=pd.DataFrame(clm3)
df3
```

```python
df3.to_csv('s.csv')
```

```python
path4='/content/pulse/data/map/user/hover/country/india/state/'
Agg_map_s4=os.listdir(path4)
Agg_map_s4
```

```python
clm4={'state':[],'year':[],'quter':[],'district':[],'registeredUsers':[],'appOpens':[]}
for s4 in Agg_map_s4:
  s4_4=path4+s4+'/'
  Agg_y4=os.listdir(s4_4)
  Agg_y4
  for y4 in Agg_y4:
    y4_4=s4_4+y4+'/'
    Agg_j4=os.listdir(y4_4)
    Agg_j4
    for j4 in Agg_j4:
      j4_4= y4_4+j4
      data4=open(j4_4,'r')
      d4=json.load(data4)
      for z4 in d4['data']['hoverData']:
        district=z4
        registeredUsers=d4['data']['hoverData'][z4]['registeredUsers']
        appOpens=d4['data']['hoverData'][z4]['appOpens']
        clm4['district'].append(district)
        clm4['registeredUsers'].append(registeredUsers)
        clm4['appOpens'].append(appOpens)
        clm4['state'].append(s4)
        clm4['year'].append(y4)
        clm4['quter'].append(int(j4.strip('.json')))
df4=pd.DataFrame(clm4)
df4
```

```python
df4.to_csv('z4.csv')
```

```python
path5='/content/pulse/data/top/transaction/country/india/state/'
Agg_s5=os.listdir(path5)
Agg_s5
```

```python
path6='/content/pulse/data/top/user/country/india/state/'
Agg_s6=os.listdir(path6)
Agg_s6
```

```python
clm6={'state':[],'year':[],'quter':[],'district_name':[],'registeredUsers':[]}
for s6 in Agg_s6:
  s6_6=path6+s6+'/'
  Agg_y6=os.listdir(s6_6)
  for y6 in Agg_y6:
    y6_6=s6_6+y6+'/'
    Agg_j=os.listdir(y6_6)
    for j in Agg_j:
      j_j=y6_6+j
      dat=open(j_j,'r')
      dd=json.load(dat)
      for h in dd['data']['districts']:
          Name=h['name']
          user=h['registeredUsers']
          clm6['district_name'].append(Name)
          clm6['registeredUsers'].append(user)
          clm6['state'].append(s6)
          clm6['year'].append(y6)
          clm6['quter'].append(int(j.strip('.json')))
df6=pd.DataFrame(clm6)
df6
```
